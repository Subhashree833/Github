#Importing Required Libraries

**Reading dataset**

```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/train.csv")
df
```

```python
df.head()
```

# Checking Missing values

```python
df.isnull().sum()
```

## Dataset Information

```python
df.info()
```

# Feature and Target Selection

```python
X=df.drop('SalePrice',axis=1)
y=df['SalePrice']
```

```python
print("Shape of x:",X.shape)
print("Shape of y:",y.shape)
```

# Train Test Split

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)
```

```python
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
```

```python
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

```python
df.describe()
```

```python
import pandas as pd

X_train = pd.get_dummies(X_train)
X_test  = pd.get_dummies(X_test)
```

```python
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
import numpy as np

X_train = np.nan_to_num(X_train)
X_test  = np.nan_to_num(X_test)
```

# Model Training using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

# Model Evaluation(before prediction)

```python
lr.score(X_test,y_test)
```

# Model Prediction

```python
lr.predict([X_test[0,:]])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred=lr.predict(X_test)
X_pred=lr.predict(X_train)
y_pred
X_pred
```

```python
print("MAE Value:",mean_absolute_error(y_test,y_pred))
print("MSE Value:",mean_squared_error(y_test,y_pred))
```

```python
lr_rmse= np.sqrt(mean_squared_error(y_test,y_pred))
lr_rmse
```

**Accuracy**

```python
lr_r2 = r2_score(y_test,y_pred)
lr_r2
```

#Traning model using Random Forest

```python
print("Training Score:",lr.score(X_train,y_train))
print("Testing Score:",lr.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
```

**Accuracy**

```python
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_model.predict(X_test)))
rf_r2   = r2_score(y_test, rf_model.predict(X_test))
rf_rmse
rf_r2
```

```python
print("Linear Regression RMSE:", lr_rmse)
print("Linear Regression R2:", lr_r2)

print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2:", rf_r2)
```

```python
import numpy as np
import pandas as pd
import gradio as gr
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/train.csv")

# Select important features
features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath']
df = df[features + ['SalePrice']]

# Drop missing values
df = df.dropna()

# Split features and target
X = df[features]
y = df['SalePrice']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction function
def predict_price(OverallQual, GrLivArea, GarageCars, TotalBsmtSF, FullBath):
    input_data = np.array([[OverallQual, GrLivArea, GarageCars, TotalBsmtSF, FullBath]])
    prediction = model.predict(input_data)
    return f"🏠 Predicted House Price: ₹ {round(prediction[0], 2)}"

# Create UI
demo = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Slider(1, 10, label="Overall Quality (1-10)"),
        gr.Number(label="Above Ground Living Area (sq ft)"),
        gr.Slider(0, 4, label="Garage Capacity (cars)"),
        gr.Number(label="Total Basement Area (sq ft)"),
        gr.Slider(0, 4, label="Number of Full Bathrooms")
    ],
    outputs="text",
    title="🏠 House Price Prediction System",
    description="Enter house details to estimate the selling price."
)

demo.launch()
```

```python
import numpy as np
import pandas as pd
import gradio as gr
from sklearn.linear_model import LinearRegression

# Load Dataset


df = pd.read_csv("/content/train.csv")

features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath']
df = df[features + ['SalePrice']]
df = df.dropna()

X = df[features]
y = df['SalePrice']

model = LinearRegression()
model.fit(X, y)

# Prediction Function

def predict_price(OverallQual, GrLivArea, GarageCars, TotalBsmtSF, FullBath):
    input_data = np.array([[OverallQual, GrLivArea, GarageCars, TotalBsmtSF, FullBath]])
    prediction = model.predict(input_data)
    return f"🏠 Predicted House Price: ₹ {round(prediction[0], 2)}"


# ===============================
# WORKING CSS FOR GRADIO
# ===============================
css = """
.gradio-container {
    background: url(https://lifetinyhouse.com/wp-content/uploads/2024/01/Tiny-House-Design-With-Swimming-Pool12-1024x576.png)
                no-repeat center center fixed;
    background-size: cover;
}

/* Blur overlay */
.gradio-container::before {
    content: "";
    position: fixed;
    inset: 0;
    background: rgba(0,0,0,0.55);
    backdrop-filter: blur(8px);
    z-index: -1;
}

/* Floating white card */
#main-box {
    background: rgba(255,255,255,0.95);
    border-radius: 22px;
    padding: 35px;
    margin: 40px auto;
    max-width: 760px;
    box-shadow: 0 25px 60px rgba(0,0,0,0.45);
}

/* Title */
.title-text {
    text-align: center;
    color: white;
    font-size: 42px;
    font-weight: bold;
    margin-top: 25px;
}

/* Button style */
button {
  background: linear-gradient(135deg, #ff7e5f, #ff3f34) !important;
    color: white !important;
    border-radius: 14px !important;
    font-size: 18px !important;
    box-shadow: 0 8px 22px rgba(0,0,0,0.35);
}
"""

# UI

with gr.Blocks(css=css) as demo:

    gr.Markdown("<div class='title-text'>🏠 House Price Prediction System</div>")

    with gr.Column(elem_id="main-box"):

        OverallQual = gr.Slider(1, 10, label="Overall Quality (1-10)")
        GrLivArea = gr.Number(label="Above Ground Living Area (sq ft)")
        GarageCars = gr.Slider(0, 6, label="Garage Capacity (cars)")
        TotalBsmtSF = gr.Number(label="Total Basement Area (sq ft)")
        FullBath = gr.Slider(0, 4, label="Number of Full Bathrooms")

        btn = gr.Button("✨ Predict Price")

        output = gr.Textbox(label="Estimated Price")

        btn.click(
            predict_price,
            inputs=[OverallQual, GrLivArea, GarageCars, TotalBsmtSF, FullBath],
            outputs=output
        )

demo.launch()
```
